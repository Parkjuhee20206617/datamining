# sorting 
iloc: 인덱스(숫자)
loc: 열이름으로

```python
import pandas as pd
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
#title columns만 sorting(오름차순)
movies.title.sort_values() 
#movies['title'].sort_values()
```

```python
movies.title.sort_values(ascending=False)#내림차순으로 sorting
```

```python
# nothing changed /  inplace 필요(원본은 갱신되지 않음)
movies.head()
```

```python
movies.sort_values(['content_rating', 'duration']).head() #columns 두개 sorting 가능함
```

```python
# filter rows by column value
# 상영 시간이 200분이상인 영화 목록 정보를 보여라(duration>200)
```

```python
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
booleans[0:5]
```

```python
is_long = pd.Series(booleans)  # series로 변환
is_long.head()
```

```python
movies[is_long].head() 
```

```python
# for loop 없이
is_long = movies.duration >= 200 # series operation의 output은 series
is_long.head()
#movies[movies.duration >= 200] 더짧게
```

```python
movies[is_long].head()
```

```python
movies.loc[[0,1,2], :] # 0,1,2 행, 전체 열
# movies.loc[[0:2], :] # same
# movies.loc[0:2]# same, bu not recommended since not explicit
```

```python
movies.loc[:, ['title', 'duration']].head() #전체에서 title, duration까지 출력
```

```python
movies.loc[0:2, 'title': 'duration']
```

```python
movies.head(3).drop(['star_rating','actors_list'], axis=1) #열삭제
```

```python
movies[movies.genre == 'Crime'].head()
#movies.loc[movies.genre == 'Crime', :].head()
#movies.loc[movies.genre == 'Crime']
```

```python
movies.loc[movies.genre == 'Crime', 'title'] # 장르가 crime인 것의 title가져오기
#movies.loc[movies.genre == 'Crime'].title
```

```python
# movies[movies.duration >= 200].genre
movies.loc[movies.duration >= 200,'genre'] # 기간이 200이상인 장르 가져오기기
```

```python
movies.iloc[:, [0,3]].head(3) # 열은 인덱스0,3
```

```python
movies.iloc[:, 0:4].head()  # loc과는 다름 4는 포함하지 않음
```

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
(movies.duration >= 200) & (movies.genre == 'Drama')  # inside
```
