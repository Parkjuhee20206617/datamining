```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec
```

```python
x = np.linspace(0,np.pi*2 ,100)
fig = plt.figure()
#plt.style.use('default')
plt.plot(x,np.sin(x), 'r-')
plt.plot(x , np.cos(x),'b:')

fig.savefig('sin_cos.png') # 그림 저장
```

```python
#도전문제
from matplotlib.gridspec import GridSpec
fig, ax = plt.subplots(3,3) #3행3열
gs = GridSpec(nrows=3, ncols=3) #그리드 사이즈 조절

x = np.random.randn(100)
y = np.random.randn(100)
ax0 = fig.add_subplot(gs[0, :])
ax0.scatter(x,y) # 산점도

x = np.linspace(0,np.pi*2 ,100)
ax1=fig.add_subplot(gs[1,:2])
plt.plot(x, 'r--') #선그래프

x = np.arange(10)
y = np.random.uniform(1,10,10)
ax4 = fig.add_subplot(gs[2,1])
ax4.bar(x,y) #막대그래프

x = np.linspace(0,10,100)
y = np.cos(x)
ax3 = fig.add_subplot(gs[2, 0])
ax3.plot(x,y,color='green') #cos그래프

z=np.random.uniform(0,1,(9,5))
ax2 = fig.add_subplot(gs[1:,2])
ax2.imshow(z) #히트맵
```

```python
x = np.arange(5)
txts = np.loadtxt('table-1.txt',delimiter=',', dtype=np.int32)
#(파일 항상 저장해야함)텍스트 불러오기- ','으로 구분되어 있음
years = txts[0]
domestic = txts[1]
foreign = txts[2]
print(txts)
```

```python
#도전문제
plt.barh(x+0.4,domestic,height=0.3) #x+0.4만큼 bar띄움, height는 두께정함
plt.barh(x,foreign,height=0.3) # 수평막대그래프
plt.yticks(x,years) 
```

```python
plt.bar(x+0.4,domestic,width=0.3) #x+0.4만큼 bar띄움, width는 두께정함
plt.bar(x,foreign,width=0.3) # 수평막대그래프
plt.xticks(x,years)
```

# 히트맵 : 열의 분포 형태로 정보 표현

```python
#연습문제1
x = np.linspace(0,np.pi,10) # 곡선을 그릴 때는 linspace.
y = np.sin(x)
n = np.zeros((10,10)) #격자 10x10

for i in range(10):
  for j in range(10):
    n[i,j] = (np.sin(x[i]) + np.sin(x[j]))/2
plt.imshow(n)
```

```python
#연습문제2
x = np.linspace(0,np.pi,100)
y = np.sin(x)
n = np.zeros((100,100))

for i in range(100):
  for j in range(100):
    n[i,j]= np.sin(x[j]) # i: 가로(행)로 나옴, j: 세로(열)로 나옴
plt.imshow(n)
```

```python
#연습문제3
n =1000
x = np.random.randn(n) #표준정규분포
y = np.random.randn(n)
plt.scatter(x,y,color='green',alpha=0.2) #투명도조절
```

```python
#연습문제4
x = np.arange(2023,2030)
y1 = np.array([1,3,7,5,9,7,14])
y2 = np.arange(1,14,2)
fig,ax1 = plt.subplots()

ax1.bar(x, y2, color='pink',label='demand')
ax1.set_xlabel('year')
ax1.set_ylabel('price')
ax1.legend(loc='upper right')
ax1.set_title('exercise')
ax1.set_xticks([2024,2026,2028])
ax1.set_yticks([0,5,10,15,20])

ax2 = ax1.twinx()
ax2.plot(x,y1,color='green',label='price',marker='s',alpha=0.5)
ax2.set_ylabel('demand')
ax2.set_yticks([0,5,10,15,20])
ax2.legend(loc='upper left')
```
