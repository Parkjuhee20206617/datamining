# duplicate(중복행), drop(제거)

```python
# 6.14 duplicate
# 라이브러리 불러오기
import pandas as pd

# 중복 데이터를 갖는 데이터프레임 만들기
df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
df
print('\n')

# 데이터프레임 전체 행 데이터 중에서 중복값 찾기
df_dup = df.duplicated()
print(df_dup)
print('\n')

# 데이터프레임의 특정 열 데이터에서 중복값 찾기
col_dup = df['brand'].duplicated()
print(col_dup)
```

```python
# 데이터프레임에서 중복 행을 제거
df2 = df.drop_duplicates()
print(df2)
print('\n')

#  중복 행을 제거
df3 = df.drop_duplicates(subset=['brand']) #중복행(지정) 삭제: brand가 같은 행 삭제
print(df3)
print('\n')

# 중복 행을 제거
df4 = df.drop_duplicates(subset=['brand','style'], keep='last') 
print(df4)
```

```python
weather = pd.read_csv('https://github.com/dongupak/DataML/raw/main/csv/weather.csv',encoding='CP949') 
weather['month'] = pd.DatetimeIndex(weather['일시']).month 
#시계열(DatetimeIndex): 시간의 흐름에 따라 순차적으로 관측한 값들의 집합

monthly = [None for x in range(12)]
monthly_wind = [0 for x in range(12)]
for i in range(12):
  monthly[i] = weather [weather['month'] == i + 1] #달별로 불리
  monthly_wind[i] = monthly[i].mean()['평균기온'] # 개별 데이터 평균 분석(평균풍속속)

months = np.arange(1,13)
plt.bar(months, monthly_wind, color='green')
plt.xlabel('month')
plt.ylabel('temperature')
```

# 테이블 데이터의 결합: concat과 merge

```python
# concat: 데이터를 하나로 합치는 방법
df_1 = pd.DataFrame( {'A' : ['a10', 'a11', 'a12'], 
                      'B' : ['b10', 'b11', 'b12'],
                      'C' : ['c10', 'c11', 'c12']} , 
                      index = ['가', '나',  '다'] )

df_2 = pd.DataFrame( {'B' : ['b23', 'b24', 'b25'],
                      'C' : ['c23', 'c24', 'c25'],
                      'D' : ['d23', 'd24', 'd25']} , 
                      index = ['다', '라',  '마'] )
df_1,df_2
```

```python
df_3 = pd.concat( [df_1, df_2]) 
df_3
```

```python
df_3_1 = pd.concat([df_1, df_2], axis=0, join='outer')
#axis=0 :기본 , join : 두개 'outer'(합집합:기본) or 'inner'(교집합) 선택
df_3_1
```

```python
df_4 = pd.concat( [df_1, df_2], join='inner')  #inner : 교집합(공통인 것만 출력)
print(df_4)
```

```python
print('left outer\n', df_1.merge(df_2 , how ='left',on='B')) # df_1기준
print('\nright outer\n', df_1.merge(df_2 , how ='right',on='B')) # df_2기준
print('\nfull outer\n', df_1.merge(df_2 , how ='outer',on='B')) # 합집함
print('\ninner\n', df_1.merge(df_2 , how ='inner',on='B')) # 교집합
```

```python
df_3 = df_1.merge(df_2, how = 'outer', 
                  left_index = True, right_index = True )
print(df_3)
```
