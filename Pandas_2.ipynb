```python
# rename columns
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.index #인덱스
ufo.columns  #열이름
```

```python
ufo.rename(columns = {'Colors Reported':'Colors_Reported', 'Shape Reported':'Shape_Reported'},inplace=True) # 열이름 다시 지정, 원본에 갱신
```

```python
ufo.columns
```

```python
ufo_cols = ['city', 'colors reported', 'shape reported', 'state','time'] #rename 말고 다르게 지정
ufo.columns = ufo_cols
ufo.head()
```

```python
# tips
ufo.columns = ufo.columns.str.replace(' ', '_') #빈칸 -> _로 변경
ufo.columns
```

#Drop

```python
ufo.drop('colors_reported', axis=1, inplace=True) #colors_reported삭제
# ufo.drop('colors_reported', axis='columns', inplace=True) # index or columns
ufo.head()
```

```python
ufo.drop(['city', 'state'], axis=1, inplace=True) #삭제후 원본 갱신
#ufo.drop([0,1], axis=0, inplace=True) #0,1행 삭제후 원본갱신
ufo.head()
```
